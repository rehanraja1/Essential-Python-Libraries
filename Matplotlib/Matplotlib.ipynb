**Pie**

1.   **Used to Show Proportions**


**Attributes**

*   lables
*   startangle
*   autospct
*   shadow
*   explode




```python
nums=[10,20,30,40,50]
plt.figure(figsize=(8,7),dpi=90)
plt.pie(nums)
plt.show()
```

```python
fruits = [
    "Apple",
    "Banana",
    "Mango",
    "Orange",
    "Grapes"
]
plt.figure(figsize=(8,7),dpi=130)
plt.pie(nums,labels=fruits)
plt.show()
```

```python
plt.figure(figsize=(8,7),dpi=130)
plt.pie(nums,labels=fruits,autopct='%1.2f%%',explode=[0,0,0.1,0,0])
plt.show()
```

```python
plt.figure(figsize=(8,7),dpi=130)
plt.pie(nums,labels=fruits,autopct='%1.2f%%',explode=[0,0,0.6,0.1,0])
plt.show()
```

```python
plt.figure(figsize=(8,7),dpi=130)
plt.pie(nums,labels=fruits,autopct='%1.2f%%',explode=[0,0,0.1,0,0],shadow=True)
plt.show()
```

```python
plt.figure(figsize=(8,7),dpi=130)
plt.pie(nums,labels=fruits,autopct='%1.2f%%',explode=[0,0,0.1,0,0],startangle=180)
plt.show()
```

**Box Plot**


1.   Used for Numerical Feature
2.   Used to Detect Outlier
3.   shows qartile values





```python
salaries = [
    40000,
    46000,
    52000,
    58000,
    65000,
    72000,
    80000,
    90000,
    105000,
    250000   # Outlier
]
```

```python
plt.figure(figsize=(10,4),dpi=90)
plt.boxplot(salaries)
plt.show()
```

```python
import numpy as np
print('Q1: ',np.quantile(salaries,0.25))
print('Q2: ',np.quantile(salaries,0.50))
print('Q2: ',np.quantile(salaries,0.75))
```

```python
plt.figure(figsize=(20,10),dpi=300)
plt.subplot(2,2,1)
plt.plot(years_of_exp,salaries,'-r')

plt.subplot(2,2,2)
plt.scatter(years_of_exp,salaries)

plt.subplot(2,2,3)
plt.boxplot(salaries)

plt.subplot(2,2,4)
plt.violinplot(salaries)
```

```python
plt.figure(figsize=(20,6),dpi=300)
plt.subplot(1,2,1)
plt.plot(years_of_exp,salaries,'r-')

plt.subplot(1,2,2)
plt.scatter(years_of_exp,salaries)
```
